```python
#2. replace rm na's with mean of the column
lin_data['rm']=lin_data['rm'].fillna(lin_data['rm'].mean())

#No null values now
lin_data.isna().any()
```

```python
lin_data.columns.tolist
```

```python
#check the columns with their unique counts
#lin_data['crim'].nunique() #not categorical

def check_if_numeric_feature_is_categorical(df,threshold=5):
    total_columns=df.columns.values.tolist()
    categorical_fields=[]
    for col in total_columns:
        uniq_count=df[col].nunique() 
        if uniq_count<=threshold:
            categorical_fields.append(col)
    return categorical_fields
```

```python
cat_columns=check_if_numeric_feature_is_categorical(df=lin_data)
print(cat_columns)
```

```python
#3. preprocessing
columns_to_standard_scale=['crim', 'zn', 'indus', 'nox',  'dis', 'rad', 'tax', 'ptratio', 'b', 'lstat','rm']
columns_to_minmaxscale=['age']

standard_scaler=StandardScaler()
min_max_scaler=MinMaxScaler()

standard_scaler.fit(lin_data[columns_to_standard_scale])
min_max_scaler.fit(lin_data[columns_to_minmaxscale])

lin_data[columns_to_standard_scale]=standard_scaler.transform(lin_data[columns_to_standard_scale])
lin_data[columns_to_minmaxscale]=min_max_scaler.transform(lin_data[columns_to_minmaxscale])
```

```python
lin_data['medv_log_transformed']=np.log(lin_data['medv'])
```

```python
X=lin_data.drop(['medv','medv_log_transformed'],axis=1)
y=lin_data['medv_log_transformed']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
#Custom Linear Regressor
lr_model_custom=LR(learning_rate=0.1)
```

```python
lr_model_custom.fit(X_train,y_train)
```

```python
y_predicted=lr_model_custom.predict(X_test)
```

```python
lr_model_custom.mse(y_test,y_predicted)
```

```python
lr_model_custom.rmse(y_test,y_predicted)
```

```python
r_squared, adj_r_squared = lr_model_custom.r_squared(X_test, y_test,y_predicted)
```
